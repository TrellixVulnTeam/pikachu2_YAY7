```python
import pandas as pd
from tqdm import tqdm
import json
import time
import gezi
```

```python
dcs = []
for k in tqdm(range(30)):
  d = gezi.read_parquet(f'../input/train/part_{k + 1}/context.parquet')
  d = d[d.label==1][['did', 'vid', 'timestamp']]
  d['day'] = k + 1
  dcs += [d]
dc = pd.concat(dcs)
```

```python
dc = dc.sort_values(['did', 'timestamp'], ascending=[True, False])
```

```python
def gtime(timestamp):
  x = time.localtime(timestamp)
  return f'{x.tm_mon}-{x.tm_mday}-{x.tm_hour}-{x.tm_min}'
```

```python
def get_dw(did):
  watches = de[de.did==did].watch.apply(json.loads).values[0]
  dw = pd.DataFrame({'timestamp':[x[0] for x in watches], 'vid': [x[1] for x in watches], 
                   'time_str': [gtime(x[0]) for x in watches]})
  return dw
```
