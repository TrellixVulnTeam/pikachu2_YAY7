```python
import pandas as pd
import gezi
from gezi import tqdm
import plotly.express as px
```

```python
dcs = []
for k in tqdm(range(30)):
  d = gezi.read_parquet(f'../input/train/part_{k + 1}/context.parquet')
  d['day'] = k + 1
  dcs += [d]
d = gezi.read_parquet('../input/eval/context.parquet')
d['day'] = 31
dcs += [d]
dc = pd.concat(dcs)
```

```python
dc_train = pd.concat(dcs[:29])
```

```python
dc_valid = dcs[29]
```

```python
new_dids = set(dc_valid.did) - set(dc_train.did)
```

```python
len(new_dids) / len(set(dc_valid.did))
```

```python
dc_train2 = pd.concat([dc_train, dc_valid])
```

```python
new_dids2 = set(dcs[-1].did) - set(dc_train2.did)
```

```python
len(new_dids2) / len(set(dcs[-1].did))
```

```python
len(dc_valid[dc_valid.did.isin(new_dids)]) / len(dc_valid)
```

```python
len(dcs[-1][dcs[-1].did.isin(new_dids2)]) / len(dcs[-1])
```
