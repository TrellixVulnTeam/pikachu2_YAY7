```python
def batch_to_numpy_images_and_labels(data):
    x, labels = data
    ids = x['id']
    images = x['image']
    numpy_images = images.numpy()
    numpy_labels = labels.numpy()
    if numpy_labels.dtype == object: # binary string in this case, these are image ID strings
        numpy_labels = [None for _ in enumerate(numpy_images)]
    # If no labels, only image IDs, return None for labels (this is the case for test data)
    return ids.numpy(), numpy_images, numpy_labels
```

```python
def display_one_image(image, title, subplot, red=False, titlesize=16):
    plt.subplot(*subplot)
    plt.axis('off')
    plt.imshow(image)
    if len(title) > 0:
        plt.title(title, fontsize=int(titlesize) if not red else int(titlesize/1.2), color='red' if red else 'black', fontdict={'verticalalignment':'center'}, pad=int(titlesize/1.5))
    return (subplot[0], subplot[1], subplot[2]+1)
```

```python
import math, re, os, time
import tensorflow as tf
import numpy as np
from collections import namedtuple
from matplotlib import pyplot as plt
def display_batch_of_images(databatch, predictions=None):
    """This will work with:
    display_batch_of_images(images)
    display_batch_of_images(images, predictions)
    display_batch_of_images((images, labels))
    display_batch_of_images((images, labels), predictions)
    """
    # data
    ids, images, labels = batch_to_numpy_images_and_labels(databatch)
    if labels is None:
        labels = [None for _ in enumerate(images)]
        
    # auto-squaring: this will drop data that does not fit into square or square-ish rectangle
    rows = int(math.sqrt(len(images)))
    cols = len(images)//rows
        
    # size and spacing
    FIGSIZE = 13.0
    SPACING = 0.1
    subplot=(rows,cols,1)
    if rows < cols:
        plt.figure(figsize=(FIGSIZE,FIGSIZE/cols*rows))
    else:
        plt.figure(figsize=(FIGSIZE/rows*cols,FIGSIZE))
    
    # display
    for i, (id, image, label) in enumerate(zip(ids, images[:rows*cols], labels[:rows*cols])):
        title = str(id)
        correct = True
        # if predictions is not None:
        #     title, correct = title_from_label_and_target(predictions[i], label)
        dynamic_titlesize = FIGSIZE*SPACING/max(rows,cols)*40+3 # magic formula tested to work from 1x1 to 10x10 images
        subplot = display_one_image(image, title, subplot, False, titlesize=dynamic_titlesize)
    
    #layout
    #plt.tight_layout()
    if label is None and predictions is None:
        plt.subplots_adjust(wspace=0, hspace=0)
    else:
        plt.subplots_adjust(wspace=SPACING, hspace=SPACING)
    plt.show()

    subplot=(rows,cols,1)
    if rows < cols:
        plt.figure(figsize=(FIGSIZE,FIGSIZE/cols*rows))
    else:
        plt.figure(figsize=(FIGSIZE/rows*cols,FIGSIZE))
    
    # display
    for i, (id, image, label) in enumerate(zip(ids, images[:rows*cols], labels[:rows*cols])):
        title = str(id)
        correct = True
        # if predictions is not None:
        #     title, correct = title_from_label_and_target(predictions[i], label)
        dynamic_titlesize = FIGSIZE*SPACING/max(rows,cols)*40+3 # magic formula tested to work from 1x1 to 10x10 images
        subplot = display_one_image(label, title, subplot, False, titlesize=dynamic_titlesize)
    
    #layout
    #plt.tight_layout()
    if label is None and predictions is None:
        plt.subplots_adjust(wspace=0, hspace=0)
    else:
        plt.subplots_adjust(wspace=SPACING, hspace=SPACING)
    plt.show()
```

```python
import sys 
import os

from absl import app, flags
FLAGS = flags.FLAGS

# flags.DEFINE_list('image_size', [256, 256], '')

import tensorflow as tf
from tensorflow.keras import backend as K
import tensorflow_io as tfio
import numpy as np

import melt as mt
#from projects.ai.naic2020_seg.src import util
#from projects.ai.naic2020_seg.src.config import *

matches = [100, 200, 300, 400, 500, 600, 700, 800]

NUM_CLASSES = len(matches)

# @tf.function
def get_seg_labels(img, num_classes=None):
  img = tf.cast(img / 100, tf.int32) - 1

  if not num_classes:
    return img

  # TODO 
  seg_labels = np.zeros((*FLAGS.image_size, num_classes))
  for c in range(num_classes):
    seg_labels[:, :, c] = (img == c).astype(int)

  seg_labels = np.reshape(seg_labels, (FLAGS.image_size[0] * FLAGS.image_size[-1], num_classes))
  return seg_labels

def decode_image(image_data):
  # image = tf.image.decode_image(image_data, channels=3)
  image = tfio.experimental.image.decode_tiff(image_data)
  image = image[:, :, :3]
  # image = tf.cast(image, tf.float32) / 127.5 - 1 
  image = tf.cast(image, tf.float32) / 255.
  image = tf.reshape(image, [*FLAGS.image_size, 3])
  return image

def decode_labels(mask):
  mask = tf.image.decode_png(mask, dtype=tf.dtypes.uint16)
  mask = tf.reshape(mask, FLAGS.image_size)
  return get_seg_labels(mask)

class Dataset(mt.Dataset):
  def __init__(self, subset='valid', **kwargs):
    super(Dataset, self).__init__(subset, **kwargs)
    assert not FLAGS.batch_parse, 'image decode must before batch'

  def parse(self, example):
    self.auto_parse(keys=['id', 'label', 'image'])
    features = self.parse_(serialized=example)

    features['image'] = decode_image(features['image'])
    if self.subset != 'test':
      features['label'] = decode_labels(features['label'])
    else:
      features['label'] = 0
    # mt.try_append_dim(features)

    x = features
    y = features['label']
    del features['label']

    self.features = features
    return x, y
```
