```python
base = root + '/16.txt'
exp = root + '/456.txt'
li_base = [line.strip().split() for line in open(base)]
li_exp = [line.strip().split() for line in open(exp)]
```

```python
len(li_base)
```

```python
base = {}
exp = {}

def calc(li, m):
  for l in li:
    pos = int(l[0])
    dur = int(l[1])
    if dur > 12 * 60 * 60:
      dur = 60
    click = int(dur > 0) 
    if pos not in m:
      m[pos] = [0, 0, 0]
    m[pos][click] += 1
    if click:
      m[pos][2] += dur
```

```python
calc(li_base, base)
calc(li_exp, exp)
```

```python
l = []
total_show_base = 0
total_show_exp = 0
total_click_base = 0
total_click_exp = 0
total_dur_base = 0
total_dur_exp = 0
for i in range(18):
  show_base = base[i][0] 
  show_base /= 3
  click_base = base[i][1] / 3 / show_base 
  show_exp = exp[i][0] 
  click_exp = exp[i][1] / show_exp
  total_show_base += show_base
  total_click_base += base[i][1] / 3
  total_show_exp += show_exp
  total_click_exp += exp[i][1]
  dur_base = base[i][2] / 3 / show_base
  dur_exp = exp[i][2] / show_exp
  total_dur_base += base[i][2] / 3
  total_dur_exp += exp[i][2]
  l.append((i, show_base, click_base, dur_base, show_exp, click_exp, dur_exp))

l.append((-1, total_show_base, total_click_base / total_show_base, total_dur_base / total_show_base, \
           total_show_exp, total_click_exp / total_show_exp, total_dur_exp / total_show_exp))
df = pd.DataFrame(l, columns=['pos', 'show_base', 'click_base', 'dur_base', 'show_exp', 'click_exp', 'dur_exp'])
```
