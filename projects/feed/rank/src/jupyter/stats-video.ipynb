```python
def cb_info(abid):
    users = set()
    cold_users = set()
    click_users = set()
    num_clicks = 0
    for l in li:
        if l[4] == abid:
            users.add(l[0])
            if l[2] != '0':
                click_users.add(l[0])
                num_clicks += 1
            if l[-1] in cb_users:
                cold_users.add(l[0])
    print(abid, 'num_users:', len(users), 'num_cold_users:', len(cold_users), 'ratio:', len(cold_users) / len(users), 'click_users:', len(click_users))
    print(num_clicks/13)
```

```python
base = {}
exp = {}

def calc(li, m):
  for l in li:
    pos = int(l[POS])
    video_time = int(l[VIDEO_TIME])
    dur = int(l[2])
    click = int(dur > 0) 
    if pos not in m:
      m[pos] = [0, 0, 0, 0, 0]
    m[pos][click] += 1
    if click:
      m[pos][2] += dur
    m[pos][3] += video_time
    m[pos][4] +=  dur / video_time
```

```python
calc(li_base, base)
calc(li_exp, exp)
```
